```python
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include='number').corr(),annot=True,cmap='coolwarm',linewidth=0.5)
plt.title("Correlation between Numeric Column")
plt.show()
```

```python
# 1st. How many Employees are in each job?
d=df['Job Title'].value_counts()
d
```

```python
plt.barh(d.index, d, height=0.6)
plt.xlabel('Counts')
plt.ylabel('Job Title')
plt.title('Employees in each job')
plt.show()
```

```python
# other example
plt.bar(d.index, d , width=0.6 )
plt.xlabel('counts')
plt.ylabel('Job Title')
plt.title('Employees in each job')
plt.xticks(rotation=45)
plt.show()
```

```python
# 2nd. Gender break-down of the staff
g=df['Gender'].value_counts()
g
```

```python
plt.pie(g, labels=g.index, autopct='%.f%%')
plt.title("Gender break-down of the staff")
plt.show()
```

```python
# 3rd. Age spread of the staff
plt.hist(df['Age'], bins=[10,20,30,40,50,60,70])
plt.xlabel('Age')
plt.ylabel('Age Bins')
plt.title('Age spread of the staff')
plt.show()
```

```python
# 4th. Which jobs pay more?
df.groupby('Job Title')['Salary'].max()
```

```python
 
df[['Job Title', 'Name','Salary']].sort_values(by= "Salary", ascending=False).groupby('Job Title').head(3)
```

```python
# 6th. Top earners in each job
df.groupby(['Job Title','Name', 'Salary'])['Salary'].sum()
```

```python
 #7th Question: Qualification vs Salary
sns.relplot(df, x='Salary', y='Education Qualification', hue='Education Qualification')
```

```python
# or other example
df['Education Qualification']=df['Education Qualification'].replace({'High School Diploma':0, 'Diploma':1, "Bachelor's Degree":2, "Master's Degree":3})
df
```

```python
plt.scatter(df['Salary'], df['Education Qualification'] , c=df['Education Qualification'], cmap='turbo', alpha=0.9)
plt.xlabel('Salary')
plt.ylabel('Education Qaulification')
plt.title('Qualification vs. Salary')
plt.show()
```

```python
# 8th Staff growth trend over time
d=df.sort_values(by='Date of Join')
d['Year']=d['Date of Join'].dt.year
d
```

```python
d1=d.groupby('Year')['Year'].count()
d1
```

```python
plt.plot(d1. index, d1,color='b', marker='*')
plt.title("Staff growth trend over time")
```

```python
# 9th Question: Leave balance analysis
d2=df.groupby(['Job Title'])['Leave Balance'].mean()
d2
```

```python
plt.barh(d2.index, d2, height=0.6)
plt.xlabel('Leave Balance')
plt.ylabel('Job Title')
plt.title('Leave balance analysis')
plt.show()
```
